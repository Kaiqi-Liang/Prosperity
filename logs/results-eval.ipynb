```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Disable scientific notation globally
mpl.rcParams['axes.formatter.useoffset'] = False
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv('./round4/round4-results.csv', delimiter=";")
```

```python
df[df['product'] == "AMETHYSTS"].head(10)
```

```python
df[df['profit_and_loss']<0]
```

```python
#  Graph the 'profit_and_loss' for each unique product
for product in df['product'].unique():
    df[df['product'] == product].plot(x='timestamp', y='profit_and_loss', title=product, figsize=(3, 2))
```

```python
df
```

```python
# Plot the 'bid_price_1' and 'ask_price_1' for each unique product as boxplots
filtered_df = df
for product in filtered_df['product'].unique():
    filtered_df[filtered_df['product'] == product].plot(x='timestamp', y=['bid_price_1'], title=product)
```

```python
df[df["product"] == "STARFRUIT"]["mid_price"].mean()
```

```python
# Plot the midmarket (average of 'bid_price_1' and 'ask_price_1') for each unique product
df['midmarket'] = (df['bid_price_1'] + df['ask_price_1']) / 2
for product in df['product'].unique():
    plot = df[df['product'] == product].plot(x='timestamp', y='midmarket', title=product, figsize=(3, 2))
```

```python
# Sum all the 'profit_and_loss' for each product
df.groupby('product')['profit_and_loss'].sum()
```

```python
# plot the sum of 'profit_and_loss' for each timestamp
df.groupby('timestamp')['profit_and_loss'].sum().plot(figsize=(10, 5))
```
