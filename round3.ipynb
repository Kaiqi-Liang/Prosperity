```python
import pandas as pd
```

```python
DATA_FOLDER = "./round-3-island-data-bottle"

day0prices_df = pd.read_csv(f"{DATA_FOLDER}/prices_round_3_day_0.csv", index_col=0, delimiter=";")
day0trades_df = pd.read_csv(f"{DATA_FOLDER}/trades_round_3_day_0_nn.csv", index_col=0, delimiter=";")
```

```python
df = day0trades_df
# Go through all symbols and get their min, max, mean, median, std
for symbol in df["symbol"].unique():
  curr_df = df[df["symbol"] == symbol]
  print(f"Symbol: {symbol}")
  print(f"  Min price: {curr_df['price'].min()}")
  print(f"  Max price: {curr_df['price'].max()}")
  print(f"  Mean price: {curr_df['price'].mean()}")
  print(f"  Median price: {curr_df['price'].median()}")
  print(f"  Std price: {curr_df['price'].std()}")
```
